```python
>>> from galpy.potential import MiyamotoNagaiPotential, NFWPotential, HernquistPotential
import matplotlib.pyplot as plt
>>> mp= MiyamotoNagaiPotential(a=0.5,b=0.0375,normalize=.6)
>>> np= NFWPotential(a=4.5,normalize=.35)
>>> hp= HernquistPotential(a=0.6/8,normalize=0.05)
>>> from galpy.potential import plotRotcurve
>>> plotRotcurve([hp,mp,np],Rrange=[0.01,20.],grid=1001,yrange=[0.,1.2])
>>> mp.plotRotcurve(Rrange=[0.01,10.],grid=1001,overplot=True,color='k',linestyle='--')
>>> hp.plotRotcurve(Rrange=[0.01,10.],grid=1001,overplot=True,color='r')
>>> np.plotRotcurve(Rrange=[0.01,10.],grid=1001,overplot=True,color='b')
plt.show()
```

```python
>>> from galpy.orbit import Orbit
>>> mp= MiyamotoNagaiPotential(a=0.5,b=0.0375,amp=1.,normalize=1.)
>>> o= Orbit.from_name('Omega cen', ro=8., vo=420.)
>>> import numpy
>>> ts= numpy.linspace(0,10,1000)
>>> o.integrate(ts,mp,method='odeint')
>>> o.plot3d(d1='x',d2='y',d3='z')
plt.show()
```
